```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#常數
N0 = 6e23
k = 1.38e-23
ε_values = [1e-21, 2e-21, 3e-21]

#溫度範圍
T = np.linspace(1, 400, 1000)

def H(T, ε):
    return N0 * ε - N0 * k * T * np.log(1 + np.exp(-ε / (k * T)))

def S(T, ε):
    return ((N0 * ε) / T) * np.exp(-ε / (k * T)) / (1 + np.exp(-ε / (k * T))) + N0 * k * np.log(1 + np.exp(-ε / (k * T)))

def U(T, ε):
    Z = np.exp(-ε / (k * T)) * (1 + np.exp(-ε / (k * T)))
    return N0 * ε * ((1 + Z * np.exp(-ε / (k * T))) / (1 + np.exp(-ε / (k * T))))

def Cv(T, ε):
    return ((N0 * ε**2) / (k * T**2)) * (np.exp(-ε / (k * T)) / (1 + np.exp(-ε / (k * T)))**2)

def plot(T, ε_values, function, ylabel, title):
    for ε in ε_values:
        plt.plot(T, function(T, ε), label=f'ε = {ε}J')
    plt.title(title)
    plt.xlabel('Temperature T (K)')
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.legend()
    plt.show()

plot(T, ε_values, H, 'Helmholtz Free Energy A', 'A vs T for Different ε')
plot(T, ε_values, S, 'Entropy S', 'S vs T for Different ε')
plot(T, ε_values, U, 'Internal Energy U', 'U vs T for Different ε')
plot(T, ε_values, Cv, 'Heat Capacity Cv', 'Cv vs T for Different ε')
```
